```python
# while loop
num = 1
while num <= 10:
    print(num)
    num +=1
```

```python
# break statement || This statement immediately terminates the loop it's currently in.
num = 1
while num <= 10:
  if num ==4:
    break
  print(num)
  num +=1
```

```python
#continue || skips the current iteration and moves to next iteration
num = 0
while num <= 10:
  num +=1
  if num ==3:
    continue
  print(num)
```

# `pass` Statement

`pass` is a null operation. It does nothing and is used as a placeholder where a statement is syntactically required but no action is desired.

```python
#class task || number guessing Game
Secret_num = 22
while True:
    user_num =int(input("Enter the number to match: "))
    if Secret_num == user_num:
        print("correct guess")
        break # This will exit the 'while True' loop
    else:
        print("try again")
        continue
```

```python
#creating a simple ATM menu
print("Welcome to the ATM")
balance = 1000
while True:
    print("\nMenu:")
    print("1. Check Balance")
    print("2. Deposit")
    print("3. Exit")
    choice = input("Enter your choice (1/2/3): ")
    if choice == '1':
        print(f"Your balance is ${balance}")
    elif choice == '2':
        amount = float(input("Enter the deposit amount: $"))
        balance += amount
        print(f"${amount} has been deposited. Your new balance is ${balance}")
    elif choice == '3':
      print("You exited the menu")
      break
    else:
      print( "Invalid choice. Please select a valid option (1/2/3)")
      continue
```

```python
#range (start,stop)
n = int(input("Enter a number to find it's multiplication table: "))
for i in range(1,11):
  print(f"{n}*{i} = {n*i}")
```

```python
#range(start,stop,step)
for i in range(0,11,2):
  print(i)
```

```python
#countdown
for i in range(10,0,-1):
  print(i)
```

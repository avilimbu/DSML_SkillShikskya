```python
#return largest
def largest(a,b,c):
  largest = max(a,b,c)
  return largest

largest(10,50,84)
```

```python
# return the number of vowels
def countVowels(text):
  count = 0
  vowels = "aeiou"
  for char in text:
    if char in vowels:
      count += 1

  return count

text = input("Enter the sentence/ word: ").lower()

countVowels(text)
```

```python
#enumerate
fruits = ["Apple", "Banana", "Mango"]

for index, fruit in enumerate(fruits, start=1):
    print(index, fruit)
```

```python
names = ["Richard", "David", "Robert"]

for i, name in enumerate(names, start =1):
  print(f"Name: {name}, Rollno: {i}")
```

```python
#zip
data1 = ["Richard", "David", "Robert"]
data2 = [1,2,3]

for name,rollno in zip(data1,data2):
  print(rollno,name, sep="->")
```

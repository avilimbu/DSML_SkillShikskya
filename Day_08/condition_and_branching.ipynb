```python
#if statement
age = int(input("Enter the age: "))
if age <18:
  print("You are not eligible to vote")
else:
  citizenship = input("Do you have a citizenship: ").strip().upper()
  if citizenship == "YES":
    print("You are eligible to vote")
  else:
    print("You are not eligible to vote")
```

```python
# Ternary operator
# value_if_true if_statement else value_if_false

n = int(input("enter a number: "))
print("even" if (n%2==0) else "odd")
```

```python
#class task 3

mark = int(input("enter the mark: "))
if mark <=100 and mark >=90:
  print("A+, Outstanding. And your GPA is 4.0")
elif mark <90 and mark >=80:
  print("A, Excellent. And your GPA is 3.6")
elif mark <80 and mark >=70:
  print("B+, Very Good. And your GPA is 3.2")
elif mark <70 and mark >=60:
  print("B, Good. And your GPA is 2.8")
elif mark <60 and mark >=50:
  print("C+, Average. And your GPA is 2.4")
elif mark <50 and mark >=40:
  print("c, Below Average. And your GPA is 2.0")
elif mark <40 and mark >=30:
  print("D, Poor. And your GPA is 1.6")
elif mark < 30:
  print("you are non graded:")
else:
  print("invalid input")
```

```python
#class task 3
#checking if num is positive or even
num = int(input("enter the number: "))
if num >=0 and (num%2==0):
  print("positive and even")
elif num >=0 and (num%2!=0):
  print("positive and odd")
else:
  print("you entered negative number.")
```

```python
#check if a day is saturday or sunday using or
day = input("enter the day in a week: ").strip().capitalize()
if day == "Saturday" or day == "Sunday":
  print("It is a weekend")
else:
  print("It is not a weekend")
```

```python
#any function with if
is_admin = True
is_mentor = False

if any([is_admin,is_mentor]):
  print("You are logged in")
```

```python
#all() function with if
is_admin = True
is_mentor = False

if all([is_admin,is_mentor]):
  print("You are logged in")
else:
  print("You are not logged in")
```

```python
#all() function with if
marks = [45,67,90]

if all([mark >=35 for mark in marks]):
  print("All students are passed")
else:
  print("Not all the students scored the passing mark")
```

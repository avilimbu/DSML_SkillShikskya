```python
#*args
def summ(*num):
    total = 0

    for i in num:
        total +=i

    return total


num = [2, 3, 5, 6]

print(summ(*num))
```

```python
def add(*args):
    return sum(args)

print(add(3, 4, 5))
```

```python
#*kwargs
def func(**kwargs):
  for k,value in kwargs.items():
    print(k, "=",value)

func(s1='student', s2="employee", s3="ABC")
```

```python
def introduce(**kwargs):
  data = []
  for k, v in kwargs.items():
    print(k,v)
    data.append(k+":"+str(v))
  return ",".join(data)
introduce(name="avhi",course="dsml")
```

```python
def student_info(*args, **kwargs):
  print("subject: ",args)
  print("student_details: ",kwargs)

student_info(
    "math",
    "dsml",
    name="rohsna",
    id=2,
    course="dsml")
```

```python
#class task
def product_details(**kwargs):
  total =0
  for k,v in kwargs.items():
    print(k,"=",v)
    total+=v
  return total

product_details(apple=250,banana=140,face_wash=400)
```

```python
def cart_total(*args):
    total = 0

    for price in args:
        total += price

    return total


print(cart_total(100, 200, 50, 150))
```
